# Data Manipulation

```python
df['Suscribers'] = df['Suscribers'].astype(int)
df['Visits'] = df['Visits'].astype(int)
df['Likes'] = df['Likes'].astype(int)
df['Comments'] = df['Comments'].astype(int)
```

```python
df.dtypes
```

```python
df['Suscribers'] = df['Suscribers'] / 1000000
df.rename(columns={'Suscribers': 'Subscribers(in Millions)'}, inplace=True)
df['Visits'] = df['Visits'] / 100000
df.rename(columns={'Visits': 'Visits(in Lakhs)'}, inplace=True)
df['Likes'] = df['Likes'] / 100
df.rename(columns={'Likes': 'Likes(in Hundreds)'}, inplace=True)
df
```

# Handling Null values

```python
df.isnull().sum()
```

```python
mode_df=df["Categories"].mode()
mode_df
```

```python
df['Categories'].fillna(df['Categories'].mode()[0], inplace=True)
df
```

# 7)Streamers with High Preformance

```python
sub_avg #average Subscribers(in Millions)	
```

```python
high_streamers=df[df["Subscribers(in Millions)"]>sub_avg]
high_streamers
```

```python
visit_avg # average Visits(in Lakhs)
```

```python
high_Visits=df[df["Visits(in Lakhs)"]>visit_avg]
high_Visits
```

```python
likes_avg # average Likes(in Hundreds)
```

```python
high_Likes=df[df["Likes(in Hundreds)"]>likes_avg]
high_Likes
```

```python
com_avg # average Comments
```

```python
high_Comments=df[df["Comments"]>com_avg]
high_Comments
```

# Top Perfroming streamers

```python
merged_df1 = pd.merge(high_streamers, high_Visits, on='Rank', how='inner')
merged_df1
```

```python
top_performance=pd.merge(merged_df1,high_Likes,on='Rank',how='inner')
top_performance
```
